```python
import GEOparse
import pandas as pd

# Define the GEO accession number
geo_id = "GSE135304"

# Download the dataset, it will be saved in the current directory
gse = GEOparse.get_GEO(geo=geo_id, destdir="./")

# Print general information about the dataset
print(gse)

# Get the sample information (e.g., series matrix table)
gse_matrix = gse.pivot_samples('VALUE')
```

```python
first_sample = list(gse.gsms.keys())[0]  # Get the first sample
print(gse.gsms[first_sample].table.head())  # Display the table for the first sample
```

```python
# Save as CSV file
gse_matrix.to_csv('GSE135304.csv', index=False)

# Verify file is saved
!ls -l
```

```python
# Access metadata of the dataset
print(gse.metadata)

# Explore the first sample to see its metadata and expression data
first_sample = list(gse.gsms.keys())[0]  # Get the first sample
gse.gsms[first_sample].metadata   # Metadata of the first sample
gse.gsms[first_sample].table.head()  # Expression data (first few rows)
```

```python
import pandas as pd

# Combine expression data from all samples into a DataFrame
expression_data = pd.DataFrame({gsm_name: gsm.table['VALUE'] for gsm_name, gsm in gse.gsms.items()})

# Set the gene probe identifiers as the index (ID_REF)
expression_data.index = gse.gsms[first_sample].table['ID_REF']

# Transpose the data so that samples are rows and gene probes (ID_REF) are columns
transposed_expression_data = expression_data.T

# Save the transposed dataset as a CSV file
transposed_expression_data.to_csv('Transposed_GSE135304_Expression_Data.csv', index=True)

# Verify the transposed dataset format
transposed_expression_data.info()
```

```python
# Initial dataset details
print("Initial dataset details:")
print(f"Number of samples: {data.shape[0]}")
print(f"Number of features (including 'SampleID' and 'TargetLabel'): {data.shape[1]}")
print(f"Number of features (genes only): {data.shape[1] - 2}\n")  # Excluding SampleID and TargetLabel

# Exclude 'SampleID' and 'TargetLabel' columns
data_features = data.drop(columns=['SampleID', 'TargetLabel'])

# Set a threshold for outlier detection (1.5, 2, or 3 times the IQR)
outlier_threshold = 3  # Adjust this value as needed (e.g., 1.5, 2, or 3)

# Calculate Q1, Q3, and IQR for each feature
Q1 = data_features.quantile(0.25)
Q3 = data_features.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers for each feature using the chosen threshold
outliers = (data_features < (Q1 - outlier_threshold * IQR)) | (data_features > (Q3 + outlier_threshold * IQR))

# Get column names with any outliers
columns_with_outliers = outliers.any(axis=0)

# Print columns with outliers
print(f"Number of columns with outliers (threshold = {outlier_threshold} * IQR): {columns_with_outliers.sum()}")
print("Columns with outliers:")
print(columns_with_outliers[columns_with_outliers].index.tolist(), "\n")

# Drop columns with outliers from the entire dataset
data_no_outliers = data.drop(columns=columns_with_outliers[columns_with_outliers].index)

# Final dataset details
print("Dataset details after removing columns with outliers:")
print(f"Number of samples: {data_no_outliers.shape[0]}")
print(f"Number of features (including 'SampleID' and 'TargetLabel'): {data_no_outliers.shape[1]}")
print(f"Number of features (genes only): {data_no_outliers.shape[1] - 2}")
```
